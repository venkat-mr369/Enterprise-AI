# 01 - Data Analysis
Exploratory analysis of the raw customer churn dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/customer_churn.csv')
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
df['Churn'].value_counts(normalize=True).plot(kind='bar', title='Churn Rate')
plt.show()
```

```python
df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack().plot(kind='bar', stacked=True)
plt.title('Churn by Contract Type')
plt.show()
```
